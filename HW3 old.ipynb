### Задание 1

Напишите программу, которая по данному числу n от 1 до 9 выводит на экран n пингвинов. Изображение одного пингвина имеет размер 5 × 9 символов, между двумя соседними пингвинами также имеется пустой (из пробелов) столбец

```python
def penguin(n):
    n = int(n)
    hat = n * "   _~_   "
    eyes = "  (o o)  " * n
    body = " /  V  \ " * n
    body2 = "/(  _  )\\" * n
    leg = "  ^^ ^^  " * n
    print(hat)
    print(eyes)
    print(body)
    print(body2)
    print(leg)
   
```

```python
penguin(input())
```
